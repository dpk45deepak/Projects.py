**Customer Survey Intelligence**


```python
import pandas as pd

df = pd.read_csv("../data/dataset.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution of Customers")
plt.show()
```

"The age distribution shows a strong concentration between 21 and 26 years, indicating that online food delivery services are predominantly used by young adults, particularly students and early-career professionals."

```python
plt.figure(figsize=(8,5))
df['Occupation'].value_counts().plot(kind='bar')
plt.title("Occupation Distribution")
plt.xticks(rotation=45)
plt.show()
```

Students are the primary customer segment for food delivery platforms.

"The occupation analysis reveals that students constitute the largest share of online food delivery users, highlighting them as the primary target segment for food delivery platforms."


```python
plt.figure(figsize=(10,5))
df['Monthly Income'].value_counts().plot(kind='bar')
plt.title("Income Distribution of Customers")
plt.xticks(rotation=45)
plt.show()
```

Most food delivery users are financially dependent individuals.

"The income distribution reveals that a large proportion of online food delivery users have no personal income, reinforcing the finding that students constitute the dominant customer segment."

#### WHY People Use Online Food Delivery


```python
reasons = [
    "Ease and convenient",
    "Time saving",
    "More restaurant choices",
    "Easy Payment option"
]

for col in reasons:
    print("\n", col)
    print(df[col].value_counts())
```

"The primary drivers of online food delivery adoption are convenience and time efficiency, with over 75% of respondents strongly agreeing that ease of use is the most important factor. Payment convenience, while relevant, plays a comparatively smaller role in influencing customer decisions."

```python
plt.figure(figsize=(4,4))
df['Output'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Customer Satisfaction Level")
plt.ylabel("")
plt.show()
```

"The satisfaction analysis indicates that nearly 78% of customers are satisfied with online food delivery services, demonstrating strong customer acceptance. However, a notable minority of dissatisfied users suggests opportunities for service quality improvements."
